```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.ndimage import distance_transform_edt
import cv2
import tensorflow as tf
from collections import deque
import random

class ImageEnvironment:
    def __init__(self, edge_points, initial_com, ideal_com,prev_dist):
        self.edge_points = edge_points
        self.num_points = len(edge_points)
        self.ideal_com = ideal_com
        self.weights = np.ones(self.num_points)  # Initialize weights to 1
        self.com = initial_com  # Initialize COM as [x, y]
        self.prev_dist=prev_dist

    def reset(self):
        """Reset the environment to its initial state."""
        self.weights = np.ones(self.num_points)
        self.com = np.ones(2)

    def update_com(self):
        """Calculate the center of mass (COM) of the edge points."""
        total_weight = np.sum(self.weights)
        if total_weight == 0:
            self.com = np.zeros(2)
        else:
            weighted_sum = np.sum(np.array(self.edge_points) * self.weights[:, np.newaxis], axis=0)
            self.com = weighted_sum / total_weight

    def take_action(self, point_idx, com_color='red'):
        """Take an action to add weight to a specified edge point."""
        self.weights[point_idx] += 1
        self.update_com()
        # Check if the added point already exists in edge_points
        added_point = self.edge_points[point_idx]
        if added_point not in self.edge_points:
            self.edge_points.append(added_point)  # Add the point to edge_points
            self.num_points += 1  # Update the number of points
            self.weights = np.append(self.weights, [1])  # Add a default weight for the new point
        self.visualize(point_idx, com_color)  # Visualize the environment after taking the action
        return self.get_state()  # Return the next state after taking the action

    def get_reward(self):
        """Calculate the reward based on the change in Manhattan distance of the COM from the ideal position."""
        # print("Ideal COM ",self.ideal_com)
        # print("Actual COM ",self.com)
        # dist = np.sqrt((new_x-ideal_center_x)**2+(new_y-ideal_center_y)**2) * (1/(dump_points[i][0]))
        current_distance = np.linalg.norm(self.com - self.ideal_com, ord=1)
        if current_distance < self.prev_dist:
            # print("Current distance = ",current_distance)
            # print("Prev distance = ",np.linalg.norm(self.ideal_com, ord=1))
            reward = abs(1000/current_distance)  # Reward for moving towards the ideal position
        else:
            reward = -abs(np.log(current_distance))  # Penalty for moving away from the ideal position

        # print("Distance bw ideal and actual COM = ",current_distance)
        # print("Reward = ", reward)
        self.prev_dist = current_distance
        return reward

    def get_state(self):
        """Return the current state of the environment."""
        return self.weights.copy(), self.com

    def visualize(self, point_idx, com_color='red'):
        """Visualize the environment with the edge points, weights, and the current center of mass."""
        plt.figure(figsize=(8, 6))
        plt.scatter([p[0] for p in self.edge_points], [p[1] for p in self.edge_points], color='black', label='Edge Points')
        plt.scatter(self.com[0], self.com[1], color=com_color, marker='o', label='Updated COM')  # Plot updated COM
        plt.scatter(self.ideal_com[0], self.ideal_com[1], color='blue', marker='*', label='Ideal COM')  # Plot ideal COM
        plt.scatter(self.edge_points[point_idx][0], self.edge_points[point_idx][1], color='red', marker='s', label='Added Weight')  # Plot point where weight is added
        plt.annotate(f'({self.edge_points[point_idx][0]}, {self.edge_points[point_idx][1]})', (self.edge_points[point_idx][0], self.edge_points[point_idx][1] + 10), color='red')  # Annotate point where weight is added
        plt.text(self.com[0], self.com[1] + 10, f'Updated COM: ({self.com[0]}, {self.com[1]})', ha='center', color=com_color)  # Print updated COM
        plt.xlabel('X Coordinate')
        plt.ylabel('Y Coordinate')
        plt.title('Image Environment')
        plt.legend()
        plt.grid(True)
        plt.gca().set_aspect('equal')  # Set aspect ratio to equal to maintain correct proportions
        plt.gca().invert_yaxis()  # Reverse the y-axis
        plt.show()

# Deep Q-Network (DQN)
class DQN(tf.keras.Model):
    def __init__(self, state_size, action_size):
        super(DQN, self).__init__()
        self.dense1 = tf.keras.layers.Dense(64, activation='relu')
        self.dense2 = tf.keras.layers.Dense(64, activation='relu')
        self.output_layer = tf.keras.layers.Dense(action_size,activation='sigmoid')

    def call(self, inputs):
        x = self.dense1(inputs)
        x = self.dense2(x)
        return self.output_layer(x)

# Replay Buffer
class ReplayBuffer:
    def __init__(self, capacity):
        self.buffer = deque(maxlen=capacity)

    def add(self, experience):
        self.buffer.append(experience)

    def sample(self, batch_size):
        return random.sample(self.buffer, batch_size)

# DQN Agent
class DQNAgent:
    def __init__(self, state_size, action_size, epsilon=1, epsilon_decay=0.990, epsilon_min=0.01, gamma=0.95, batch_size=32):
        self.state_size = state_size
        self.action_size = action_size
        self.epsilon = epsilon
        self.epsilon_decay = epsilon_decay
        self.epsilon_min = epsilon_min
        self.gamma = gamma
        self.batch_size = batch_size
        self.model = DQN(state_size, action_size)
        self.target_model = DQN(state_size, action_size)
        self.target_model.set_weights(self.model.get_weights())
        self.optimizer = tf.keras.optimizers.Adam(learning_rate=0.001)
        self.replay_buffer = ReplayBuffer(capacity=10000000)

    def select_action(self, state):
        weights, com = state
        flat_state = np.concatenate((weights, com))
        print(self.epsilon)
        if np.random.rand() <= self.epsilon:
            return random.randrange(self.action_size)
        q_values = self.model.predict(np.expand_dims(flat_state, axis=0))[0]
        # print(q_values)
        return np.argmax(q_values)

    def train(self):
        print(len(self.replay_buffer.buffer))
        if len(self.replay_buffer.buffer) < self.batch_size:
            return
        batch = self.replay_buffer.sample(self.batch_size)
        states, actions, rewards, next_states = zip(*batch)
        # Concatenate weights and COM tuples into flat arrays
        states = np.concatenate([np.concatenate(state) for state in states], axis=0).reshape(self.batch_size, -1)
        next_states = np.concatenate([np.concatenate(next_state) for next_state in next_states], axis=0).reshape(self.batch_size, -1)
        actions = np.array(actions)
        rewards = np.array(rewards)
        target_q_values = self.target_model.predict(next_states)
        target_q_values = rewards + self.gamma * np.amax(target_q_values, axis=1)
        q_values = self.model.predict(states)
        q_values[np.arange(self.batch_size), actions] = target_q_values
        with tf.GradientTape() as tape:
            loss = tf.reduce_mean(tf.square(q_values - self.model(states)))
        grads = tape.gradient(loss, self.model.trainable_variables)
        self.optimizer.apply_gradients(zip(grads, self.model.trainable_variables))

    def update_target_network(self):
        self.target_model.set_weights(self.model.get_weights())

# Load the image (replace 'filled_image_1' with your image path)
image = filled_image_1

# Detect edges using Canny edge detection
edges = cv2.Canny(image, 100, 200)

# Find contours from the edges
contours, _ = cv2.findContours(edges, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Sample points along the contours
edge_points = []
for contour in contours:
    points = contour[:, 0, :]
    edge_points.extend(points)

# Convert edge_points to list of tuples (x, y)
edge_points = [(p[0], p[1]) for p in edge_points]

# Define the initial and ideal COM positions (replace with your desired positions)
initial_com = np.array([507, 396])
ideal_com = np.array([457.5 ,389.5])

# Create the environment
env = ImageEnvironment(edge_points, initial_com, ideal_com,prev_dist=1e9+7)

state_size = env.num_points * 2  # Number of edge points * 2 (x and y coordinates)
action_size = env.num_points  # Number of edge points (actions correspond to selecting which pixel to add weight to)

# Create the DQN agent
agent = DQNAgent(state_size, action_size)

distances = []

# Train the agent
episodes = 250
for episode in range(episodes):
    state = env.get_state()
    c = 0
    done = False
    while not done:
        c += 1
        action = agent.select_action(state)
        next_state = env.take_action(action)  # Take the selected action
        reward = env.get_reward()
        agent.replay_buffer.add((state, action, reward, next_state))
        agent.train()
        agent.update_target_network()
        state = next_state
        if c > 50:
            break
        if done:
            break
        # Calculate the distance between the ideal COM and the current COM
        current_distance = np.linalg.norm(env.com - env.ideal_com, ord=1)
        distances.append(current_distance)

    # Decay epsilon
    if agent.epsilon > agent.epsilon_min:
        agent.epsilon *= agent.epsilon_decay
    
    agent.replay_buffer.buffer.clear()


# main_model_weights = agent.model.get_weights()
# target_model_weights = agent.target_model.get_weights()

# print("Main Model Weights:")
# for layer_weights in main_model_weights:
#     print(layer_weights.shape)  # Print the shape of each layer's weights

# print("\nTarget Model Weights:")
# for layer_weights in target_model_weights:
#     print(layer_weights.shape)  # Print the shape of each layer's weights
    

# Plot the distance between the ideal COM and the current COM over iterations
plt.figure(figsize=(8, 6))
plt.plot(range(len(distances)), distances, marker='o', linestyle='-')
plt.xlabel('Iterations')
plt.ylabel('Distance between Ideal COM and Current COM')
plt.title('Distance between Ideal COM and Current COM over Iterations')
plt.grid(True)
plt.show()

# Test the agent after training (optional)
# You can add code here to test the trained agent on the environment
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.ndimage import distance_transform_edt
import cv2
import tensorflow as tf
from collections import deque
import random

class ImageEnvironment:
    def __init__(self, edge_points, initial_com, ideal_com,prev_dist):
        self.edge_points = edge_points
        self.num_points = len(edge_points)
        self.ideal_com = ideal_com
        self.weights = np.ones(self.num_points)  # Initialize weights to 1
        self.com = initial_com  # Initialize COM as [x, y]
        self.prev_dist=prev_dist

    def reset(self):
        """Reset the environment to its initial state."""
        self.weights = np.ones(self.num_points)
        self.com = np.ones(2)

    def update_com(self):
        """Calculate the center of mass (COM) of the edge points."""
        total_weight = np.sum(self.weights)
        if total_weight == 0:
            self.com = np.zeros(2)
        else:
            weighted_sum = np.sum(np.array(self.edge_points) * self.weights[:, np.newaxis], axis=0)
            self.com = weighted_sum / total_weight

    def take_action(self, point_idx, com_color='red'):
        """Take an action to add weight to a specified edge point."""
        self.weights[point_idx] += 1
        self.update_com()
        # Check if the added point already exists in edge_points
        added_point = self.edge_points[point_idx]
        if added_point not in self.edge_points:
            self.edge_points.append(added_point)  # Add the point to edge_points
            self.num_points += 1  # Update the number of points
            self.weights = np.append(self.weights, [1])  # Add a default weight for the new point
        self.visualize(point_idx, com_color)  # Visualize the environment after taking the action
        return self.get_state()  # Return the next state after taking the action

    def get_reward(self):
        """Calculate the reward based on the change in Manhattan distance of the COM from the ideal position."""
        # print("Ideal COM ",self.ideal_com)
        # print("Actual COM ",self.com)
        # dist = np.sqrt((new_x-ideal_center_x)**2+(new_y-ideal_center_y)**2) * (1/(dump_points[i][0]))
        current_distance = np.linalg.norm(self.com - self.ideal_com, ord=1)
        if current_distance < self.prev_dist:
            # print("Current distance = ",current_distance)
            # print("Prev distance = ",np.linalg.norm(self.ideal_com, ord=1))
            reward = abs(1000/current_distance)  # Reward for moving towards the ideal position
        else:
            reward = -abs(np.log(current_distance))  # Penalty for moving away from the ideal position

        # print("Distance bw ideal and actual COM = ",current_distance)
        # print("Reward = ", reward)
        self.prev_dist = current_distance
        return reward

    def get_state(self):
        """Return the current state of the environment."""
        return self.weights.copy(), self.com

    def visualize(self, point_idx, com_color='red'):
        """Visualize the environment with the edge points, weights, and the current center of mass."""
        plt.figure(figsize=(8, 6))
        plt.scatter([p[0] for p in self.edge_points], [p[1] for p in self.edge_points], color='black', label='Edge Points')
        plt.scatter(self.com[0], self.com[1], color=com_color, marker='o', label='Updated COM')  # Plot updated COM
        plt.scatter(self.ideal_com[0], self.ideal_com[1], color='blue', marker='*', label='Ideal COM')  # Plot ideal COM
        plt.scatter(self.edge_points[point_idx][0], self.edge_points[point_idx][1], color='red', marker='s', label='Added Weight')  # Plot point where weight is added
        plt.annotate(f'({self.edge_points[point_idx][0]}, {self.edge_points[point_idx][1]})', (self.edge_points[point_idx][0], self.edge_points[point_idx][1] + 10), color='red')  # Annotate point where weight is added
        plt.text(self.com[0], self.com[1] + 10, f'Updated COM: ({self.com[0]}, {self.com[1]})', ha='center', color=com_color)  # Print updated COM
        plt.xlabel('X Coordinate')
        plt.ylabel('Y Coordinate')
        plt.title('Image Environment')
        plt.legend()
        plt.grid(True)
        plt.gca().set_aspect('equal')  # Set aspect ratio to equal to maintain correct proportions
        plt.gca().invert_yaxis()  # Reverse the y-axis
        plt.show()

# Deep Q-Network (DQN)
class DQN(tf.keras.Model):
    def __init__(self, state_size, action_size):
        super(DQN, self).__init__()
        self.dense1 = tf.keras.layers.Dense(64, activation='relu')
        self.dense2 = tf.keras.layers.Dense(64, activation='relu')
        self.output_layer = tf.keras.layers.Dense(action_size,activation='sigmoid')

    def call(self, inputs):
        x = self.dense1(inputs)
        x = self.dense2(x)
        return self.output_layer(x)

# Replay Buffer
class ReplayBuffer:
    def __init__(self, capacity):
        self.buffer = deque(maxlen=capacity)

    def add(self, experience):
        self.buffer.append(experience)

    def sample(self, batch_size):
        return random.sample(self.buffer, batch_size)

# DQN Agent
class DQNAgent:
    def __init__(self, state_size, action_size, epsilon=1, epsilon_decay=0.990, epsilon_min=0.01, gamma=0.95, batch_size=32):
        self.state_size = state_size
        self.action_size = action_size
        self.epsilon = epsilon
        self.epsilon_decay = epsilon_decay
        self.epsilon_min = epsilon_min
        self.gamma = gamma
        self.batch_size = batch_size
        self.model = DQN(state_size, action_size)
        self.target_model = DQN(state_size, action_size)
        self.target_model.set_weights(self.model.get_weights())
        self.optimizer = tf.keras.optimizers.Adam(learning_rate=0.001)
        self.replay_buffer = ReplayBuffer(capacity=10000000)

    def select_action(self, state):
        weights, com = state
        flat_state = np.concatenate((weights, com))
        print(self.epsilon)
        if np.random.rand() <= self.epsilon:
            return random.randrange(self.action_size)
        q_values = self.model.predict(np.expand_dims(flat_state, axis=0))[0]
        # print(q_values)
        return np.argmax(q_values)

    def train(self):
        print(len(self.replay_buffer.buffer))
        if len(self.replay_buffer.buffer) < self.batch_size:
            return
        batch = self.replay_buffer.sample(self.batch_size)
        states, actions, rewards, next_states = zip(*batch)
        # Concatenate weights and COM tuples into flat arrays
        states = np.concatenate([np.concatenate(state) for state in states], axis=0).reshape(self.batch_size, -1)
        next_states = np.concatenate([np.concatenate(next_state) for next_state in next_states], axis=0).reshape(self.batch_size, -1)
        actions = np.array(actions)
        rewards = np.array(rewards)
        target_q_values = self.target_model.predict(next_states)
        target_q_values = rewards + self.gamma * np.amax(target_q_values, axis=1)
        q_values = self.model.predict(states)
        q_values[np.arange(self.batch_size), actions] = target_q_values
        with tf.GradientTape() as tape:
            loss = tf.reduce_mean(tf.square(q_values - self.model(states)))
        grads = tape.gradient(loss, self.model.trainable_variables)
        self.optimizer.apply_gradients(zip(grads, self.model.trainable_variables))

    def update_target_network(self):
        self.target_model.set_weights(self.model.get_weights())

# Load the image (replace 'filled_image_1' with your image path)
image = filled_image_1

# Detect edges using Canny edge detection
edges = cv2.Canny(image, 100, 200)

# Find contours from the edges
contours, _ = cv2.findContours(edges, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Sample points along the contours
edge_points = []
for contour in contours:
    points = contour[:, 0, :]
    edge_points.extend(points)

# Convert edge_points to list of tuples (x, y)
edge_points = [(p[0], p[1]) for p in edge_points]

# Define the initial and ideal COM positions (replace with your desired positions)
initial_com = np.array([507, 396])
ideal_com = np.array([457.5 ,389.5])

# Create the environment
env = ImageEnvironment(edge_points, initial_com, ideal_com,prev_dist=1e9+7)

state_size = env.num_points * 2  # Number of edge points * 2 (x and y coordinates)
action_size = env.num_points  # Number of edge points (actions correspond to selecting which pixel to add weight to)

# Create the DQN agent
agent = DQNAgent(state_size, action_size)

distances = []

# Train the agent
episodes = 250
for episode in range(episodes):
    state = env.get_state()
    c = 0
    done = False
    while not done:
        c += 1
        action = agent.select_action(state)
        next_state = env.take_action(action)  # Take the selected action
        reward = env.get_reward()
        agent.replay_buffer.add((state, action, reward, next_state))
        agent.train()
        agent.update_target_network()
        state = next_state
        if c > 50:
            break
        if done:
            break
        # Calculate the distance between the ideal COM and the current COM
        current_distance = np.linalg.norm(env.com - env.ideal_com, ord=1)
        distances.append(current_distance)

    # Decay epsilon
    if agent.epsilon > agent.epsilon_min:
        agent.epsilon *= agent.epsilon_decay
    
    agent.replay_buffer.buffer.clear()


# main_model_weights = agent.model.get_weights()
# target_model_weights = agent.target_model.get_weights()

# print("Main Model Weights:")
# for layer_weights in main_model_weights:
#     print(layer_weights.shape)  # Print the shape of each layer's weights

# print("\nTarget Model Weights:")
# for layer_weights in target_model_weights:
#     print(layer_weights.shape)  # Print the shape of each layer's weights
    

# Plot the distance between the ideal COM and the current COM over iterations
plt.figure(figsize=(8, 6))
plt.plot(range(len(distances)), distances, marker='o', linestyle='-')
plt.xlabel('Iterations')
plt.ylabel('Distance between Ideal COM and Current COM')
plt.title('Distance between Ideal COM and Current COM over Iterations')
plt.grid(True)
plt.show()

# Test the agent after training (optional)
# You can add code here to test the trained agent on the environment
```
